```python
from river import compose
from river import preprocessing
from river import linear_model
import itertools
from river import stream

data_stream = stream.iter_csv('data/air_pollution_dataset_modified.csv', target='pred_pollution',
                              converters={'current_pollution': float, 'dew': float,
                                          'temp': float, 'press': float,
                                          'wnd_spd': float, 'snow': float,
                                          'rain': float, 'pred_pollution': float})

# Show data stream
for x, y in itertools.islice(data_stream, 2):
    print(x, y)

# Define the pipeline
model = compose.Pipeline(
    (compose.Select('wnd_dir') | preprocessing.OneHotEncoder()) + compose.Select('current_pollution', 'dew', 'temp', 'press', 'wnd_spd', 'snow', 'rain'),
    # preprocessing.StandardScaler(),
    linear_model.LinearRegression()  # Apply linear regression
)

x, y = next(data_stream)
print(model.debug_one(x))

# Train the model on the first 100 samples from the data stream
for x, y in itertools.islice(data_stream, 20):
    y_pred = model.predict_one(x)
    model.learn_one(x, y)
    print(model.debug_one(x))
```
